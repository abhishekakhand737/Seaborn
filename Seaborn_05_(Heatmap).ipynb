## Seaborn (Heatmap)

A Seaborn Heatmap is a data visualization technique used to display data in a matrix (table) format where individual values are represented by different colors. The color intensity indicates the magnitude of the values, which helps in easily identifying patterns, relationships, and variations in the dataset.

```python
import seaborn as sns 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

### Heatmap 2D numpy array 

```python
array_2d = np.linspace(1,5,12).reshape(4,3) # numpy 2D array
print(array_2d) # numpy array
```

```python
sns.heatmap(array_2d)
```

###  Heatmap using DataFrame

```python
globalWarming_df = pd.read_csv("C:\\Users\\abhishek\\Downloads\\Who_is_responsible_for_global_warming.csv")
globalWarming_df.head()
```

### drop() method.

```python
globalWarming_df  = globalWarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
globalWarming_df
```

### create a seaborn heatmap

```python
plt.figure(figsize=(17,10))
sns.heatmap(globalWarming_df)
```

### vmin & vmax parameter

```python
plt.figure(figsize=(17,10))
sns.heatmap(globalWarming_df, vmin = 0, vmax = 19)
```

### Heatmap color change using cmap parameter.

```python
plt.figure(figsize=(17,10))
sns.heatmap(globalWarming_df, cmap="coolwarm")
```

### annot (annotation) parameter

```python
plt.figure(figsize=(17,10)) 
sns.heatmap(globalWarming_df, annot = True)
```

### 2D Numpy array to annot parameter

```python
annot_arr = np.arange(1,13).reshape(4,3) #2D numpy array with 4 rows and 3 columns
sns.heatmap(array_2d, annot= annot_arr)
```

### fmt parameter (add text on each cell)

```python
 
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
 
sns.heatmap(array_2d, annot = annot_arr, fmt="s") 
```

### Change style & format of annot (annotate)
   

```python
plt.figure(figsize=(16,9))
 
annot_kws={'fontsize':10, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.6, 
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
 
sns.heatmap(globalWarming_df, annot = True, annot_kws= annot_kws)
```

### linewidths parameter

```python
plt.figure(figsize=(16,9)) 
sns.heatmap(globalWarming_df, linewidths=4)
```

### Correlation Heatmap

```python
globalWarming_df.corr()
```

### Correlation matrix

```python
plt.figure(figsize=(17,10))
ax = sns.heatmap(globalWarming_df.corr(),annot=True, linewidth=3)
ax.tick_params(size=10, color='w',labelsize=10, labelcolor='w')
plt.title("heatmap of 'Who is responsible for global warming'", fontsize=25)
plt.show()

# or
# plt.figure(figsize=(17,10))
# sns.heatmap(globalWarming_df.corr(), annot = True)
```

### Lower triangle heatmap 

```python
plt.figure(figsize=(17,10))
corr_mx = globalWarming_df.corr() # correlation matrix
matrix = np.triu(corr_mx) # take upper correlation matrix
sns.heatmap(corr_mx, mask=matrix)
```

### Upper triangle heatmap 

```python
plt.figure(figsize=(16,9))
corr_mx = globalWarming_df.corr() # correlation matrix
matrix = np.tril(corr_mx) # take lower correlation matrix
sns.heatmap(corr_mx, mask=matrix)
```
