### Seaborn (Scatter Plot)

Seaborn scatter plot is used to visualize the relationship between two numerical variables by plotting data points on a Cartesian plane.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Importing Datasets

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

### Plot seaborn scatter plot using x, y, data parameters.

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df)

## or--- sns.scatterplot(x = titanic_df.age, y = titanic_df.fare)
## sns.scatterplot(x = titanic_df['age'], y = titanic_df['fare'])
```

### Hue Parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex")
```

### Hue_order parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
               hue_order= ['female', 'male'])
```

### Size parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "who")
```

### Sizes parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "who",
                sizes = (60, 400))
```

### Size_order parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "who",
                size_order=['child', 'man', 'woman'],)
```

### Style parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = "who",)
```

### Style order parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = "who", style_order=['child','woman','man'])
```

### Palette parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",
                palette="hot") # palette does not work without hue
```

### Markers parameter

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = "who", markers= ['3','1',3])
```

### Alpha parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, alpha = 0.4)
```

### Legend parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex", legend = False )
```

### Axes Parameters

```python
ax = sns.scatterplot(x = "age", y = "fare", data = titanic_df, )
 
ax.set(xlabel = "Age",
      ylabel = "Fare",
      title = "Seaborn Scatter Plot of Age and Fare")
```

### Keyword Arguments parameter 

```python
kwargs  =   {'edgecolor':"r",
             'facecolor':"k",
             'linewidth':2.7,
             'linestyle':'--',
            }
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "sex", sizes = (500, 1000), alpha = .7,  **kwargs)
```

### Scatter plot of age and fare

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set() 
titanic_df = sns.load_dataset("titanic")
 
plt.figure(figsize = (17,10)) 
sns.scatterplot(x = "who", y = "fare", data = titanic_df, hue = "alive", style = "alive", palette = "viridis",size = "who",
                sizes = (200, 500))
 
plt.title("Scatter Plot of Age and Fare", fontsize = 25) 
plt.xlabel("Age", fontsize = 20) # x-axis label
plt.ylabel("Fare", fontsize = 20) # y-axis label
plt.savefig("Scatter Plot of Age and Fare")
plt.show()
```
