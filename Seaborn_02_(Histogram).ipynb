## Seaborn (Histogram)

A Seaborn Histogram is a graphical representation used to show the distribution of numerical data. It divides the data into bins (intervals) and displays the frequency of data points in each bin using bars.

```python
import seaborn as sns
from scipy.stats import norm 
import matplotlib.pyplot as plt
```

### Import Datasets

```python
tips_df =sns.load_dataset("tips")
tips_df
```

### Plot tips_df Size Histogram 

```python
sns.histplot(tips_df["size"])
```

### Plot tips_df tip Histogram 

```python
sns.histplot(tips_df["tip"])
```

### Total_bill Histogram

```python
sns.histplot(tips_df["total_bill"])
```

### Plot Histogram of total_bill with bins parameters

```python
sns.histplot(tips_df["total_bill"], bins=55)
```

### Plot Histogram of "total_bill" with hist parameters

```python
sns.distplot(tips_df["total_bill"], hist = False)
```

### kde (kernal density estimator) parameters.

```python
sns.histplot(tips_df["total_bill"], kde=False)
```

### Plot Histogram of "total_bill" with rugplot parameters

```python
sns.distplot(tips_df["total_bill"],rug=True,)
```

### Plot Histogram of "total_bill" with fit and kde parameters

```python
sns.distplot(tips_df["total_bill"],fit=norm, kde = False)
```

### Histogram of "total_bill" with color parameters

```python
sns.histplot(tips_df["total_bill"],color="r",) # pass red color
```

### Converting Horizontaly Histogram

```python
sns.distplot(tips_df["total_bill"],vertical=True,) # Horizontal hist
```

### Histogram of "total_bill" with norm_hist parameter.

```python
sns.distplot(tips_df["total_bill"],norm_hist=True,)
```

### Histogram of "total_bill" with axlabel parameters

```python
sns.distplot(tips_df["total_bill"],axlabel="Total Bill",)
```

### Histogram of "total_bill" with label parameters

```python
sns.histplot(tips_df["total_bill"],label="Total Bill",)
plt.title("Histogram of Total Bill") # for histogram title
plt.legend()
```

### Histplot Set style and increase figure size

```python
plt.figure(figsize=(17,10)) 
sns.set() 
sns.histplot(tips_df["total_bill"],label="Total Bill",)
plt.title("Histogram of Total Bill") 
plt.legend() 
```

### Histplot Bins

```python
tips_df.total_bill.sort_values() 
```

### Modify histogram with bin. 

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,9))
sns.set()
sns.histplot(tips_df["total_bill"], bins = bins)
plt.xticks(bins) # set bins value 
plt.title("Histogram of Total Bill") 
plt.show()
```

### Change color, edge color, line width, line style, and alpha of histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(17,10))
sns.set()
sns.histplot(
    tips_df["total_bill"],
    color='#DC143C',
    edgecolor='#aaff00',
    linewidth=5,
    linestyle='--',
    alpha=0.9
)

plt.show()
```

```python
plt.figure(figsize=(17,10))
sns.set()
sns.histplot(
    tips_df["total_bill"],
    kde=True,
    color="#808080",
    edgecolor="#aaff00",
    linewidth=5,
    alpha=0.9
)
plt.show()
```

### sns distplot histogram keyword arguments

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(17,10))
sns.set()

sns.distplot(
    tips_df["total_bill"],
    kde=False,
    rug=False,
    
    hist_kws={'color':'#808080',
              'edgecolor':'#aaff00',
              'linewidth':5,
              'linestyle':'--',
              'alpha':0.9}
)

plt.show()
```

### sns distplot kde(kernel density estimate) keyword arguments

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(
    tips_df["total_bill"],
    hist_kws = {
        'color':'#808080',
        'edgecolor':'#aaff00',
        'linewidth':5,
        'linestyle':'--',
        'alpha':0.9
    },
    kde_kws = {
        'color':'#8e00ce',
        'linewidth':8,
        'linestyle':'--',
        'alpha':0.9
    }
)
plt.show()
```

### sns distplot rugplot keyword arguments

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,9))
sns.set()

sns.distplot(
    tips_df["total_bill"],
    
    hist_kws={
        'color':'#808080',
        'edgecolor':'#aaff00',
        'linewidth':5,
        'linestyle':'--',
        'alpha':0.9
    },
    
    kde_kws={
        'color':'#8e00ce',
        'linewidth':8,
        'linestyle':'--',
        'alpha':0.9
    },
    
    rug=True,
    
    rug_kws={
        'color':'#8e00ce',
        'linewidth':8,
        'linestyle':'--',
        'alpha':0.9
    }
)

plt.show()
```

### Seaborn histograms using sns.distplot() function

```python
# Plot multiple seaborn histogram in single graph
plt.figure(figsize=(16,9))
sns.distplot(tips_df["total_bill"], bins=9, label="total_bil")
sns.distplot(tips_df["tip"], bins=9, label="tip")
sns.distplot(tips_df["size"], bins=9, label = "size")
 
plt.legend()
```
