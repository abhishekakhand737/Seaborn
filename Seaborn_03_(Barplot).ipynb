## Seaborn (Barplot)

Seaborn barplot is used to show the relationship between a categorical variable and a numerical variable using rectangular bars that represent the mean value of the data.

```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Load dataset from github

```python
tips_df = sns.load_dataset("tips")
print(tips_df)
```

### sns.barplot() x, y parameters 

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
```

### Hue Parameter

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)
```

### Data parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df)
```

### Order parameter

```python
order = ['Sun', 'Thur', 'Fri', 'Sat'] 
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, order = order)
```

### Hue_order parameter  
#### Modify hue order

```python
hue_order = ['Female', 'Male']
 
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, hue_order = hue_order)
```

### Estimator parameter

```python
tips_df = sns.load_dataset("tips")
sns.barplot(x='day', y='total_bill', hue='sex',
            data=tips_df, estimator=np.max)
plt.show()
```

### ci (Confidence interval) parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, ci = 20)
```

### n_boot parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df,  n_boot=2)
```

### Orient parameter

```python
sns.barplot(y = 'day', x = 'total_bill', hue = 'sex',
           data = tips_df,)
```

```python
sns.barplot(x ='total_bill', y = 'size', hue = 'sex', data = tips_df,
           orient='h')
```

### Color parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, color="orange")
```

### Palette parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, palette="Blues")
```

### Saturation parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, saturation=.4)
```

### Errcolor parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, errcolor='0.5')
```

### errwidth parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, errwidth= 12)
```

### Capsize parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, capsize=1)
```

### Dodge parameter

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df, dodge= False)
```

### Keyword arguments Parameters

```python
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'}
 
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex',
           data = tips_df,**kwargs)
```

```python
sns.set() 
sns.barplot(x = 'day', y = 'total_bill',palette="Set2",   
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")
```

###  axes parameter 

```python
sns.set()
ax = sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =.9, linestyle = "-.", linewidth = 3,
           edgecolor = "g")
 
ax.set(title = "Barplot of Tips DataFrame",
      xlabel = "Days",
      ylabel = "Total Bill")
```

```python
sns.set()
plt.figure(figsize = (16,9))
 
sns.barplot(x = 'day', y = 'total_bill', 
           data = tips_df, alpha =1, linestyle = "-.", linewidth = 3,
           edgecolor = "k")
 
plt.title("Barplot of Days and Total Bill", fontsize = 20)
plt.xlabel("Days", fontsize = 15)
plt.ylabel("Total Bill", fontsize = 15)
 
plt.savefig("Barplot of Days and Total Bill")
plt.show()
```
